```python
import open3d as o3d
import numpy as np

inp = "output/gs_init/kitchen/input.ply"
fp1 = "output/gs_init/kitchen/point_cloud/iteration_7000/point_cloud.ply"
fp2 = "output/gs_init/kitchen/point_cloud/iteration_10000/point_cloud.ply"

inp_pcd = o3d.io.read_point_cloud(inp)
fp1_pcd = o3d.io.read_point_cloud(fp1)
fp2_pcd = o3d.io.read_point_cloud(fp2)

# color
fp1_pcd.paint_uniform_color([0, 0, 1])
fp2_pcd.paint_uniform_color([0, 1, 0])

# union
pcd = fp1_pcd + fp2_pcd

# dist
dists = np.asarray(fp1_pcd.compute_point_cloud_distance(fp2_pcd))
print(dists.mean())
ind = np.where(dists > 0.01)[0]
pcd_without = fp1_pcd.select_by_index(ind)


o3d.visualization.draw_plotly([pcd_without], mesh_show_wireframe=False)
```
